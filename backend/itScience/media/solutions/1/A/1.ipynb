```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

from sklearn.inspection import partial_dependence
from sklearn.inspection import plot_partial_dependence
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.neural_network import MLPClassifier
from sklearn.datasets.california_housing import fetch_california_housing

def generate(count):
  x = []
  y = []
  for ir in range(0, count):
    math = np.random.randint(1, 6)
    physics = np.random.randint(1, 6)
    russian = np.random.randint(1, 6)
    lgot = np.random.randint(0, 2)
    x.append([math, physics, russian, lgot])
    mathPlus = math >= 4
    physicsPlus = physics >= 4
    sumPlus = math + physics + russian >= 11
    y.append(1 if lgot == 1 or (mathPlus and physicsPlus and sumPlus) else 0)
  return np.array(x), np.array(y)

X, y = generate(200)
X_test, y_test = generate(40)
est = MLPClassifier(activation='logistic')
est.fit(X, y)
predict = est.predict(X_test)
validNumber = 0
for elem in range(len(predict)):
  num = 1 if predict[elem] >= 0.5 else 0
  if num != y_test[elem]:
    print(X_test[elem], predict[elem], y_test[elem])
  else:
    validNumber += 1
print("Accuracy:", validNumber / len(predict))
```
