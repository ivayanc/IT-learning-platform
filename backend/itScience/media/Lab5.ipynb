```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import Perceptron
```

```
def generate():
  x = np.array([[-1, 1], [0, 0], [1, -1], [1, 1]])
  y = [1, 1, 1, 0]
  return x, y

def get_grid():
  x_min, x_max = -2, 2
  y_min, y_max = -2, 2
  return np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))
```

```

X, y = generate()
classifier = SGDClassifier()
classifier.fit(X, y)
classifier.predict([[-2, -2]])
```

```
# some code to depict separating surface
xx, yy = get_grid()
predicted = classifier.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
right = 0
for lineIt in range(0, xx.shape[0]):
    for elemIt in range(0, xx.shape[1]):
        pred = 1 if (xx[lineIt][elemIt]**2 / 4 + yy[lineIt][elemIt]**2 / 4) <= 16 else 0
        if pred == predicted[lineIt][elemIt]:
            right += 1
plt.pcolormesh(xx, yy, predicted, cmap='autumn')
plt.scatter(X[:,0], X[:,1], s=100, cmap='autumn', edgecolors='black', linewidth=1.5);
```

```
X, y = generate()
perceptron = Perceptron()
perceptron.fit(X, y)
perceptron.predict([[-2, -2]])
```

```
xx, yy = get_grid()
predicted = perceptron.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
right = 0
for lineIt in range(0, xx.shape[0]):
    for elemIt in range(0, xx.shape[1]):
        pred = 1 if (xx[lineIt][elemIt]**2 / 4 + yy[lineIt][elemIt]**2 / 4) <= 16 else 0
        if pred == predicted[lineIt][elemIt]:
            right += 1
plt.pcolormesh(xx, yy, predicted, cmap='autumn')
plt.scatter(X[:,0], X[:,1], s=100, cmap='autumn', edgecolors='black', linewidth=1.5);
```

```
def generateBad():
  x = np.array([[-1, 1], [0, 0], [1, -1], [1, 1], [1.5, 1.5]])
  y = [1, 1, 1, 0, 1]
  return x, y
```

```
X, y = generateBad()
print(X)
print(X[:,0])
classifier = SGDClassifier()
classifier.fit(X, y)

valid = 0
for iter in range(len(X)):
  if classifier.predict([X[iter]]) == y[iter]:
    valid += 1

print("Accuracy is %f" %(valid / len(X)))

# some code to depict separating surface
xx, yy = get_grid()
predicted = classifier.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
plt.pcolormesh(xx, yy, predicted, cmap='autumn')
plt.scatter(X[:,0], X[:,1], s=100, cmap='autumn', edgecolors='black', linewidth=1.5);

```
